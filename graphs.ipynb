```python
# üìä Import libraries
import re
import matplotlib.pyplot as plt

# üìÇ Load your log file (Assume it's named 'train_log.txt' and uploaded to Kaggle Notebook)
log_path = '/kaggle/input/rcnnfeatureextractlog/rcnnnotebook.log'

# ‚¨áÔ∏è Initialize storage
epochs = []
train_losses = []
valid_losses = []
accuracies = []
norm_eds = []
print("RCNN")

# üìñ Regex patterns
epoch_info_pattern = re.compile(r'Epoch: \[(\d+)/\d+\]')
loss_info_pattern = re.compile(r'Train loss: ([0-9.]+), Valid loss: ([0-9.]+)')
acc_info_pattern = re.compile(r'Current_accuracy *: *([0-9.]+), Current_norm_ED *: *([0-9.]+)')

# üìã Parse log file
with open(log_path, 'r') as file:
    lines = file.readlines()
    current_epoch = None

    for line in lines:
        # Extract epoch number
        epoch_match = epoch_info_pattern.search(line)
        if epoch_match:
            current_epoch = int(epoch_match.group(1))

        # Extract train & valid loss
        loss_match = loss_info_pattern.search(line)
        if loss_match and current_epoch is not None:
            train_loss = float(loss_match.group(1))
            valid_loss = float(loss_match.group(2))
            train_losses.append(train_loss)
            valid_losses.append(valid_loss)
            epochs.append(current_epoch)

        # Extract accuracy & norm_ED
        acc_match = acc_info_pattern.search(line)
        if acc_match:
            accuracy = float(acc_match.group(1))
            norm_ed = float(acc_match.group(2))
            accuracies.append(accuracy)
            norm_eds.append(norm_ed)

# ‚úÖ Sanity check
print(f'Total epochs parsed: {len(epochs)}')
print(f'First 5 epochs: {epochs[:5]}')
print(f'First 5 train losses: {train_losses[:5]}')
print(f'First 5 accuracies: {accuracies[:5]}')

# üìà Plot 1: Accuracy vs Epoch
plt.figure(figsize=(8,5))
min_len = min(len(epochs), len(accuracies))
plt.plot(epochs[:min_len], accuracies[:min_len], marker='o', color='blue')
plt.title('Accuracy vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.show()


# üìà Plot 2: Norm_ED vs Epoch
plt.figure(figsize=(8,5))
min_len = min(len(epochs), len(norm_eds))
plt.plot(epochs[:min_len], norm_eds[:min_len], marker='o', color='green')
plt.title('Norm_ED vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Norm_ED')
plt.grid(True)
plt.show()

# üìà Plot 3: Train & Valid Loss vs Epoch
plt.figure(figsize=(8,5))
min_len_train = min(len(epochs), len(train_losses))
min_len_valid = min(len(epochs), len(valid_losses))
plt.plot(epochs[:min_len_train], train_losses[:min_len_train], marker='o', label='Train Loss', color='red')
plt.plot(epochs[:min_len_valid], valid_losses[:min_len_valid], marker='o', label='Valid Loss', color='orange')
plt.title('Train & Valid Loss vs Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
